sUMA DE LOS NUMEROS DE 1 HASTA N


```python
def suma (n):
  if n == 1:
    return 1
  else:
    return n+suma(n-1)
```

```python
time(suma (900))
```

```python
from scipy import linalg as la
from scipy import optimize
import numpy as np
import pandas as pd
```

```python
x = np.array([1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 100, 150, 200, 250, 300, 500, 700, 900])
y = np.array([4, 5, 6, 5, 7, 9, 7, 14, 14, 9, 10, 14, 100, 77, 92, 162, 158, 196, 296])
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize = (12,4))
##Ajustar los datos al modelo usando minimos cuadrados con un polinomio de primer orden
A = np.vstack([x**n for n in range(2)])
sol1, r1, rank,sv = la.lstsq(A.T,y)
y_fit1 = sum([s*x**n for n, s in enumerate(sol1)])

ax.plot(x,y, 'go', alpha = 0.5, label = 'Datos Reales')
ax.plot(x,y_fit1, 'b' ,lw = 2, label = 'Ajuste por minimos cuadrados (orden 1)')
ax.set_xlabel(r'$x$', fontsize = 18)
ax.set_ylabel(r'$y$', fontsize = 18)
ax.legend()
```
